Import libraries

```python
import pandas as pd
```

Load data

```python
df= pd.read_csv('/content/fetal_health.csv')
#Display first few rows
df.head()
```

Check for and handle missing values

```python
missing_values=df.isnull().sum()
print (missing_values)
```

Change data type of all columns

```python
# Convert all columns of the DataFrame to float data type
df = df.astype(float)
print(df)
```

Remove duplicate values

```python
df.drop_duplicates(inplace=True)
print(df)
```

Rename columns

```python
df.rename(columns={'accelerations': 'accs'}, inplace=True)
print (df)
```

Filter Data

```python
fetal_health_risk = df[df['fetal_health']>2]
print(fetal_health_risk)
```

Sort data


```python
df.sort_values(by='fetal_health',ascending=True, inplace=True)
print(df)
```

Load data

```python
# Assuming 'df' is your cleaned DataFrame
df.to_csv('transformed_fetal_health.csv', index=False)

# Use the Google Colab file download feature to download the file to your local machine
from google.colab import files
files.download('transformed_fetal_health.csv')
```
